# Bee Swarm Plot
Distibution par rapport à une classe / catégorie

```python
plt.gcf().set_size_inches(8,8) ## Changer taille graphique

sns.swarmplot(x="species", y="petal length (cm)", data=df, s=3.99715)
```

# Box Plot
Répérer les valeurs extremes (outliers - plages de valeur)

```python
plt.gcf().set_size_inches(8,8) ## Changer taille graphique

sns.boxplot(x="species", y="petal length (cm)", data=df)
```

IQR = distance entre les deux quartiles (25% - 75%)

Box = 50 % des données

Moustache = plage attendue

Barre milieu = médiane

Outliers (losange noir) = Hors de la plage de données

# Scatter Plot
Repérer les corrélations

```python
plt.plot(df["petal length (cm)"], 
         df["petal width (cm)"], 
         marker=".", 
         linestyle="none")
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)");
```

Correlation :

- Positive = proportionnelle

- Negative = Inversement propertionnelle

- Nulle = Aucune relation
